```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("zomato.csv")
df.shape
df.head()
df.columns.tolist()
df.isnull().sum()
df['rate'].value_counts().head(10)
print(df.shape)
print(df.isnull().sum())

df['rate'] = df['rate'].astype(str).str.replace('/5', '').str.strip()
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')  # 'NEW' becomes NaN
df.dropna(subset=['rate'], inplace=True)
df['rate'].dtype
```

```python
df.drop(columns=['url', 'phone', 'dish_liked', 'menu_item', 'reviews_list', 'address'], inplace=True)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str).str.replace(',', '').str.strip()
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')
```

```python
df.dropna(subset=['location', 'cuisines', 'approx_cost(for two people)', 'rest_type'], inplace=True)
```

```python
df.rename(columns={'approx_cost(for two people)': 'cost_for_two'}, inplace=True)
```

```python
print(df.shape)
print(df.isnull().sum())
```
